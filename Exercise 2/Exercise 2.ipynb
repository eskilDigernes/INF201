```python
"""Part 2"""

import csv

filename = "2021-09-14_party distribution_1_st_2021.csv"

# creating a dictionary to store the votes for each party
party_votes = {}

# open and read the csv file
with open(filename, mode='r', encoding="utf-8") as file:
    csv_reader = csv.reader(file, delimiter=';')
    # skip the first line
    next(csv_reader)

    # iterate through the rows
    for row in csv_reader:
        # get the party code
        party_code = row[6]
        # get the number of votes
        votes = int(row[12])
        # check if the party code is already in the dictionary
        if party_code in party_votes:
            # if it is, add the votes to the existing value
            party_votes[party_code] += votes
        else:
            # if not, add the party code to the dictionary and set the votes as the value
            party_votes[party_code] = votes

# sort party votes by largest number of votes
sorted_party_votes = sorted(party_votes.items(), key=lambda x: x[1], reverse=True)

# print the results in a table
SPACING ="\t\t"
print(f"Party{SPACING}Votes")
print("----------------------")
for party in sorted_party_votes:
    print(f'{party[0]}{SPACING}{party[1]}')
```

```python
"""Part 5"""
import csv

def print_party_stats(filename, num_parties):
    # creating a dictionary to store the votes for each party
    party_votes = {}

    # open and read the file csv file
    with open(filename, mode='r', encoding="utf-8") as file:
        csv_reader = csv.reader(file, delimiter=';')
        # skip the first line
        next(csv_reader)

        # iterate through the rows
        for row in csv_reader:
            # get the party code
            party_code = row[6]
            # get the number of votes
            votes = int(row[12])
            # check if the party code is already in the dictionary
            if party_code in party_votes:
                # if it is, add the votes to the existing value
                party_votes[party_code] += votes
            else:
                # if not, add the party code to the dictionary and set the votes as the value
                party_votes[party_code] = votes

    # sort party votes by largest number of votes
    sorted_party_votes = sorted(party_votes.items(), key=lambda x: x[1], reverse=True)[:num_parties]

    # calculate the total number of votes
    total_votes = sum(party_votes.values())

    SPACING ="\t\t"
    print(f"Party{SPACING}Votes{SPACING}Percentage      >=4%")
    print("----------------------------------------------------")
    for party in sorted_party_votes:
        percentage = round(party[1]/total_votes*100, 2)
        mark = '*' if percentage >= 4 else ''
        print(f"{party[0]}{SPACING}{party[1]}{SPACING}{percentage}%{SPACING}{mark}")

# once with "3"
print_party_stats("2021-09-14_party distribution_1_st_2021.csv", 3)

print("\n")

# and once with "7"
print_party_stats("2021-09-14_party distribution_1_st_2021.csv", 7)
```

```python
##########
# Task 2 #
##########

import re

# the pattern looks for two words that start with capital letters
pattern = r"([A-Z][A-Za-z]+).*?([A-Z][A-Za-z]+)"

def extract_names(sentence):
    match = re.search(pattern, sentence)
    if match:
        # return the names from the regex groups
        return match.group(1), match.group(2)
    return None

def print_friendships(sentences):
    print("      Friendships      ")
    print("-----------------------")
    for name in sentences:
        names = extract_names(name)
        if names:
            # calculate space between names so it is centred 
            total_length = len(names[0]) + len(names[1])
            spacing = (20 - total_length) // 2
            print(f"{' ' * spacing}{names[0]} - {names[1]}{' ' * spacing}")


# sample sentences
sentences = [
    "Ali and Per and friends.",
    "Kari and Joe know each other.",
    "James has known Peter since school days."
]

print_friendships(sentences)
```
